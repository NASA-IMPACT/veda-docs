## Approach

   1. Use `pystac_client` to open the STAC catalog 
   2. Iterate over collections and print the title of each collection

## Open STAC catalog

```python
from pystac_client import Client

STAC_API_URL = "https://openveda.cloud/api/stac/"
catalog = Client.open(STAC_API_URL)
```

## List collections

```python
collections = list(catalog.get_collections())
for collection in sorted(collections, key=lambda x: x.title or x.id):
    print(collection.title or collection.id)
```
